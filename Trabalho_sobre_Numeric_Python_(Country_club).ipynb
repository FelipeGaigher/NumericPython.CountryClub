```
# Gerar um Dataset: ID (Sem repetição de Valores do número de identificação)
def gerarID(TAMANHO):
  import numpy
  vetor = numpy.zeros(TAMANHO, dtype=int)
  i = 1
  vetor[0] = 0000 + numpy.random.randint(1000, 1100, 1)
  while i != TAMANHO:
    flag = True
    j = i - 1
    # termo = 20190000 + numpy.random.randint(0, 10, 1)
    termo = 0000 + numpy.random.randint(1000, 1100, 1)
    while j != -1:
      if termo == vetor[j]:
        flag = False
        break
      j -= 1
    if flag:
      vetor[i] = termo
      i += 1
  return vetor
```

```
# Gerar um Dataset: ID (Sem repetição de Valores do número de identificação)
def posicaoClube(TAMANHO):
  import numpy
  vetor1 = numpy.zeros(TAMANHO, dtype=int)
  i = 1
  vetor1[0] = numpy.random.randint(1, 101, 1)
  while i != TAMANHO:
    flag = True
    j = i - 1
    # termo = 20190000 + numpy.random.randint(0, 10, 1)
    termo =  numpy.random.randint(1, 101, 1)
    while j != -1:
      if termo == vetor1[j]:
        flag = False
        break
      j -= 1
    if flag:
      vetor1[i] = termo
      i += 1
  return vetor1
```

```
def pesquisarMatricula(_countryClub, _matriculaPesquisa):
  for _index, _competidor in enumerate(_countryClub):
    if _competidor[0] == _matriculaPesquisa:
      return _index
  return -1 # Matrícula NÃO encontrada na lista: Turma.
```

```
"""
-> MODELAGEM DAS LINHAS: Atletas no Country Club Society
-> MODELAGEM DAS COLUNAS: 
   DADOS: Número de identificação
          Posição
          Pontuação no clube
          Gênero (Masculino, Feminino)
          Modalidades (Volei, Natação, Futebol, Academia)
          Idade do competidor
          Tempo de cadastro no Country Club Society (meses)
          País de origem do competidor (Brasil, Argentina, Uruguai, Chile)
"""

LINHAS = 100
COLUNAS = 8
countryClub = numpy.zeros((LINHAS, COLUNAS), dtype=int)  
countryClub[:,0] = gerarID(LINHAS)                                      # IDENTIFICAÇÃO
countryClub[:,1] = posicaoClube(LINHAS)                                 # POSIÇÃO
countryClub[:,2] = numpy.random.randint(1, 1000, LINHAS, dtype=int)     # PONTUAÇÃO NO CLUBE
countryClub[:,3] = numpy.random.randint(1, 3, LINHAS, dtype=int)        # GÊNERO
countryClub[:,4] = numpy.random.randint(1, 5, LINHAS, dtype=int)        # MODALIDADES
countryClub[:,5] = numpy.random.randint(18, 35, LINHAS, dtype=int)      # IDADE
countryClub[:,6] = numpy.random.randint(1, 60, LINHAS, dtype=int)       # TEMPO EM MESES 
countryClub[:,7] = numpy.random.randint(1, 5, LINHAS, dtype=int)        # PAÍSES


#Criando as modalidades:

#indexModalidade1 == VOLEI
indexModalidade1 = countryClub[:,4] == 1
countryClub[indexModalidade1, 4]

#indexModalidade2 == NATAÇÃO
indexModalidade2 = countryClub[:,4] == 2
countryClub[indexModalidade2, 4]

#indexModalidade3 == FUTEBOL
indexModalidade3 = countryClub[:,4] == 3
countryClub[indexModalidade3, 4]

#indexModalidade4 == ACADEMIA
indexModalidade4 = countryClub[:,4] == 4
countryClub[indexModalidade4, 4]
```
